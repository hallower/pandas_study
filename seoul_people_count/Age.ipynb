```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_excel('data.xls', encoding='utf-8')
```

```python
df
```

```python
df.head()
```

```python
df.rename(columns = {'인구':'인구수','인구.1':'남자', '인구.2':'여자', '65세이상고령자':'65세이상'}, inplace=True)
```

```python
del df['기간']
del df['인구.3']
del df['인구.4']
del df['인구.5']
del df['인구.6']
del df['인구.7']
del df['인구.8']
df = df.drop([0])
df = df.drop([1])
df = df.drop([2])
df.head()
```

```python
pivot_df = df.pivot_table(df, index='자치구', aggfunc=np.sum)
pivot_df
```

```python
pivot_df.sort_values(by='인구수', ascending=False, inplace=True)
pivot_df.head()
```

```python
pivot_df['65세비율']=pivot_df['65세이상'] / pivot_df['인구수']
pivot_df
```

```python
target = ['65세비율', '남자','여자','인구수']
max = pivot_df[target].max()
min = pivot_df[target].min()
people_count = (pivot_df[target] - min )/(max-min)
people_count
```

```python
plt.figure(figsize = (10,10))
sns.heatmap(people_count.sort_values(by='인구수',ascending=False), annot=True, fmt='f', linewidths=.5, cmap='BuPu')
plt.title('서울 구별 인구, 인구 수 정렬')
plt.show()
```
